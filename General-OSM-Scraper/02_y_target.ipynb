# 02 · Y-Target

Extracts the amenity label for each establishment.

**Input:** `csv/00_base_data.csv`
**Output:** `csv/02_y_target.csv` — `osm_id`, `amenity_label`

```python
import pandas as pd

df = pd.read_csv("csv/00_base_data.csv")
print(f"Loaded {len(df)} records")
```

```python
# The label column from 01_identifiers already contains the amenity label
df_out = df[["osm_id"]].copy()
df_out["amenity_label"] = df["label"]

print(f"\nAmenity label distribution:")
print(df_out["amenity_label"].value_counts())
```

```python
df_out.to_csv("csv/02_y_target.csv", index=False, encoding="utf-8")
print(f"Saved {len(df_out)} records to csv/02_y_target.csv")
df_out.head()
```
